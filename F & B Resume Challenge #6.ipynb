Demographic Insights

```python
#Who prefers energy drink more?
gender_counts = merged_data['Gender'].value_counts()
gender_counts
```

```python
plt.bar(gender_counts.index, gender_counts)
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Who prefers energy drink more?')
# Display the plot
plt.show()
```

```python
#Which age group prefers energy drinks more?
age_group = merged_data['Age'].value_counts()
age_group
```

```python
plt.pie(age_group, labels=age_group.index)
plt.title('Which age group prefers energy drinks more?')
plt.show()
```

```python
#Which type of marketing reaches the most Youth
marketing = merged_data['Marketing_channels'].value_counts()
marketing
```

```python
plt.bar(marketing.index, marketing)
# Set custom x-axis ticks
plt.xticks(marketing.index, rotation='vertical')
plt.xlabel('Marketing Channels')
plt.ylabel('Count')
plt.title('Which type of marketing is popular?')
# Display the plot
plt.show()
```
